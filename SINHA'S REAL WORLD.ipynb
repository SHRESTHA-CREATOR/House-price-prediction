## TRAIN TEST 

```python
#
import numpy as np
def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    print(shuffled)
    test_set_size = int(len(data)*test_ratio)
    test_indices = shuffled[:test_set_size]
    train_indices =shuffled[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
house = strat_train_set.drop("MEDV",axis=1)
house_labels = strat_train_set["MEDV"].copy()
```

## MISSING ATTRIBUTES

```python
median = house["RM"].median()
```

```python
house["RM"].fillna(median)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy = "median")
imputer.fit(house)
```

```python
imputer.statistics_
```

```python
x= imputer.transform(house)
```

```python
house_tr=pd.DataFrame(x,columns= house.columns)
```

```python
house_tr.describe()
```

## CREATING A PIPELINE

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([('imputer' , SimpleImputer(strategy ="median")),('std_scaler',StandardScaler()),])
```

```python
house_num_tr= my_pipeline.fit_transform(house)
```

```python
house_num_tr
```

## SELECTING MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
#model=LinearRegression()
#model = DecisionTreeRegressor()
model.fit(house_num_tr, house_labels)
```

```python
some_data = house.iloc[:5]
```

```python
some_labels= house_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data) 
```

```python
model.predict(prepared_data)
```

```python
house_labels[:5]
```

## EVALUATING THE MODEL

```python
from sklearn.metrics import mean_squared_error 
house_prediction = model.predict(house_num_tr)
mse= mean_squared_error(house_labels,house_prediction)
rmse= np.sqrt(mse)
```

```python
rmse
```

## USING CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,house_num_tr,house_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores",scores)
    print("Mean",scores.mean())
    print("standard deviation",scores.std())
```

```python
print_scores(rmse_scores)
```
